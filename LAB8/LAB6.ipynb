```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KernelDensity
```

```python
df=pd.read_excel("data.xlsx")
df
```

```python
prior_prob_no = np.sum(df['buys_computer'] == 'no') / len(df)
prior_prob_yes = np.sum(df['buys_computer'] == 'yes') / len(df)
print(f'Prior Probability (buys_computer = no): {prior_prob_no:.3f}')
print(f'Prior Probability (buys_computer = yes): {prior_prob_yes:.3f}')
```

```python

# Convert 'age' values to numerical values for KDE
age_mapping = {'<=30': 1, '31...40': 2, '>40': 3}
df['age_numerical'] = df['age'].map(age_mapping)

# Separate data by class
data_no = df[df['buys_computer'] == 'no']['age_numerical'].values.reshape(-1, 1)
data_yes = df[df['buys_computer'] == 'yes']['age_numerical'].values.reshape(-1, 1)

print(data_no)

# Fit Kernel Density Estimation for each class
kde_no = KernelDensity(bandwidth=1.0, kernel='gaussian')
kde_no.fit(data_no)

kde_yes = KernelDensity(bandwidth=1.0, kernel='gaussian')
kde_yes.fit(data_yes)

# Generate test data for plotting
age_values = np.linspace(1, 3, 1000).reshape(-1, 1)

# Calculate class-conditional densities
density_no = np.exp(kde_no.score_samples(age_values))
density_yes = np.exp(kde_yes.score_samples(age_values))

# Check for zero values in class-conditional densities
zero_density_no = any(density_no == 0)
zero_density_yes = any(density_yes == 0)

print(f'Class-Conditional Density for "buys_computer = no" has zero values: {zero_density_no}')
print(f'Class-Conditional Density for "buys_computer = yes" has zero values: {zero_density_yes}')
```
