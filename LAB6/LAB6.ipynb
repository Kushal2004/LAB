```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("air_pollution_data.csv")
df
```

```python

feature1 = df.co
feature2 = df.no
# Create a scatter plot
plt.scatter(feature1, feature2, c='b', marker='o', label='Scatter Plot')

# Add labels and a title
plt.xlabel('co')
plt.ylabel('no')
plt.title('Scatter Plot of Feature 1 vs. Feature 2')

# Add a legend (if multiple datasets are plotted)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Example data (replace with your actual data)
feature1 = df.no
feature2 = df.nh3

# Reshape the data into NumPy arrays
feature1 = np.array(feature1).reshape(-1, 1)
feature2 = np.array(feature2)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(feature1, feature2, test_size=0.2, random_state=42)

# Create and fit the linear regression model
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

# Make predictions using the model
y_pred = regr.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Plot the data points and regression line
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

# Add labels and a title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear Regression')

# Show the plot
plt.show()
```
